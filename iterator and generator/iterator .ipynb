## what make a object iterable ??

- Container.__iter__()
    - returns an iterator object.
- iterator.__next__()
    - returns the next iterm for container. 
    - If there is no further item, then raise the StopIteration exception

```python
# iter without for loop
looper1 = iter(username)
while True:
    try:
        user = next(looper1)
        print(user)
    except StopIteration:
        break
```

```python
import itertools
```

```python
ranks = list(range(2,11)) + ['J','Q','K','A']
ranks = [str(rank) for rank in ranks]
print(ranks)
```

```python
suits = ['Hearts','Spaces','Diamond','Ace']
```

```python
decks = [card for card in itertools.product(ranks,suits)]
```

```python
for (index,cards) in enumerate(decks):
    print(1+index,cards)
```

```python
[hand for hand in itertools.combinations(decks,5)]
```
